# Bisection Method

```python
def f(x):
    return x**3 - x - 1
a = 1
b = 2
if f(a) * f(b) >= 0:
    print("Invalid interval")
else:
    print("Iteration\t a\t\t b\t\t c\t\t f(c)")
    tolerance = 0.00001
    iteration = 1
    while (b - a) / 2 > tolerance:
        c = (a + b) / 2
        fc = f(c)
        print(f"{iteration}\t\t {a:.2f}\t\t {b:.2f}\t\t {c:.2f}\t\t {fc:.2f}")
        if fc == 0:
            break
        elif f(a) * fc < 0:
            b = c
        else:
            a = c
        iteration += 1
    root = (a + b) / 2
    print("\nFinal Root = {:.2f}".format(root))
```

# False Position method

```python
def f(x):
    return x**3 - 7*x + 4
a = 0
b = 1
if f(a) * f(b) >= 0:
    print("Invalid interval")
else:
    print("Iteration\t a\t\t b\t\t c\t\t f(c)")
    tolerance = 0.00001
    iteration = 1
    while True:
        c = (((a * f(b)) - (b * f(a))) / (f(b) - f(a)))
        fc = f(c)
        print(f"{iteration}\t\t {a:.2f}\t\t {b:.2f}\t\t {c:.2f}\t\t {fc:.2f}")
        if abs(fc) < tolerance:
            break
        if f(a) * fc < 0:
            b = c
        else:
            a = c
        iteration += 1
    print("\nFinal Root = {:.2f}".format(c))
```

# Newton Raphson Method

```python
def f(x):
    return x**3 - 6*x + 4
def df(x):
    return 3*x**2 - 6
def newton_raphson(x0, tol=0.001, max_iter=100):
    for i in range(1, max_iter+1):
        x1 = x0 - f(x0)/df(x0)
        print(f"Iteration {i}: x = {x1:.3f}, f(x) = {f(x1):.3f}, difference = {abs(x1 - x0):.3f}")
        if abs(x1 - x0) < tol:
            print(f"\nRoot found: {x1:.3f}")
            return x1
        x0 = x1
    print(f"\nApproximate root after {max_iter} iterations: {x1:.3f}")
    return x1
x0 = 0.5
newton_raphson(x0)
```

# Secant method

```python
def f(x):
    return 2*x**3 - 2*x - 5
x1 = 2.0
x2 = 1.0
tolerance = 0.001
iteration = 1
print(f"{'Iteration':<10}{'x1':<10}{'x2':<10}{'x3':<10}{'f(x3)':<10}")
while True:
    f1 = f(x1)
    f2 = f(x2)
    x3 = (f2*x1 - f1*x2) / (f2 - f1)
    f3 = f(x3)
    print(f"{iteration:<10}{x1:<10.4f}{x2:<10.4f}{x3:<10.4f}{f3:<10.4f}")
    if abs(x3 - x2) < tolerance:
        print(f"\nRoot of the equation is approximately: {x3:.3f}")
        break
    x1 = x2
    x2 = x3
    iteration += 1
```

# Forward Interpolation

```python
x=[1,2,3,4,5]
y=[1,8,27,64,125]
n=len(x)
d=[y.copy()]
for i in range(1,n):
    t=[]
    for j in range(n-i):
        t.append(d[i-1][j+1]-d[i-1][j])
    d.append(t)
h=x[1]-x[0]
p=(1.7-x[0])/h
f=1
s=y[0]
for i in range(1,n):
    f=f*(p-(i-1))/i
    s=s+f*d[i][0]
print("Value of y at x = 1.7 :",s)
```

# Backward Interpolation

```python
x=[1,2,3,4,5]
y=[1,8,27,64,125]
n=len(x)
d=[y.copy()]
for i in range(1,n):
    t=[]
    for j in range(n-i):
        t.append(d[i-1][j+1]-d[i-1][j])
    d.append(t)
h=x[1]-x[0]
p=(4.7-x[-1])/h
f=1
s=y[-1]
for i in range(1,n):
    f=f*(p+(i-1))/i
    s=s+f*d[i][-1]
print("y at x=4.7 =",s)
```
