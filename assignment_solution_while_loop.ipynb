# 1 no : Display numbers from -100 to -10 using while loop

### Using While Loop

```python
i = -100
while i <= -10:
    print(i, end=" ")
    i += 1
```

# 2 no : Write a program to sum all prime numbers within a range 10 to 1000

### Using While Loop

```python
num = 10
total = 0

while num <= 1000:
    i = 2
    while i < num:
        if num % i == 0:  # if divisible  not prime
            break
        i += 1
    if i == num:  # means we never found a divisor
        total += num
    num += 1

print("Sum of primes:", total)
```

# 3 no : Find the factorial of a n! (hint, n = 7; 7 * 6 * 5 * 4 * 3 * 2 * 1)

```python
n = 7
fact = 1
i = 1
while i <= n:
    fact *= i
    i += 1
print("Factorial:", fact)
```

# 4 no : Reverse a given integer number 27956240710

```python
n = 27956240710
r = 0
while n > 0:
    d = n % 10       # get last digit
    r = r * 10 + d   # add digit to reversed number
    n //= 10         # remove last digit
print("Reversed:", r)
```
