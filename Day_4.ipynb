# Conditional Statement

# if...else

```python
x = 50
y = 100

if x > y:
    print('x is greater than y')
if y > x:
    print('y is greater than x')
    print('python conditional statement')
else:
    print('out!')
```

# if...elif...else

```python
x = 20
y = 100

if x > y:
    print('x is greater than y')
elif y > x:
    print('y is greater than x')
    print('python conditional statement')
    if x != y:
        print('not equal')
else:
    print('out!')
```

```python
x = 20
y = 100

if x > y:
    print('x is greater than y')
elif y > x:
    print('y is greater than x')
    print('python conditional statement')
    if x == y:
        print('not equal')
else:
    print('out!')
```

# Divisible or not

```python
num = float(input('Enter a number : '))

if num % 2 == 0:
    if num % 5 == 0:
        print('divisible by 5 and 2')
    else:
        print('No')
else:
    print('no')
    
```
